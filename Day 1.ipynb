# Loading data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder as le
l=le()
```

```python
data=pd.read_csv("Housing.csv")
data.head()
```

```python
data['mainroad']=l.fit_transform(data['mainroad'])
data['guestroom']=l.fit_transform(data['guestroom'])
data['basement']=l.fit_transform(data['basement'])
data['hotwaterheating']=l.fit_transform(data['hotwaterheating'])
data['airconditioning']=l.fit_transform(data['airconditioning'])
data['prefarea']=l.fit_transform(data['prefarea'])
data['furnishingstatus']=l.fit_transform(data['furnishingstatus'])
data.head()
```

```python
data.isna().sum()
```

```python
data.describe()
```

# Step 2

```python
x=data.iloc[:,1:2]
x
```

```python
y=data.iloc[:,:1]
y
```

```python
import sklearn as sk
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as lr
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=0.2,random_state=5)
```

```python
xtrain
```

```python
lin=lr()
lin.fit(xtrain,ytrain)
ypred=lin.predict(xtest)
ypred
```

# Step 3

```python
from sklearn.metrics import r2_score,mean_squared_error,explained_variance_score,mean_absolute_error
```

```python
print("Mean Squared error :",mean_squared_error(ytest,ypred))
print("\nMean Absolut error :",mean_absolute_error(ytest,ypred))
print("\nRMSE :",np.sqrt(mean_squared_error(ytest,ypred)))
print("\nR Square :",r2_score(ytest,ypred))
print("\nVariance Score :",explained_variance_score(ytest,ypred))
```

```python
b=r2_score(ytest,ypred)
print(b)
if b==1:
    print("perfect fit")
elif b>0:
    print("good fit")
else:
    print("poor fit")
    
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as lr
from sklearn.metrics import r2_score,explained_variance_score,mean_squared_error,mean_absolute_error
mtrain,mtest,ntrain,ntest=tts(m,n,test_size=0.1,random_state=6)
line=lr()
line.fit(mtrain,ntrain)
npred=line.predict(mtest)
print("Mean Squared error :",mean_squared_error(ntest,npred))
print("\nRSME :",np.sqrt(mean_squared_error(ntest,npred)))
print("\nVariance Error : ",explained_variance_score(ntest,npred))
print("\nR square: ",r2_score(ntest,npred))
print("\nMAE :",mean_absolute_error(ntest,npred))
```
