```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler,LabelEncoder


# Data Preprocessing
# Separate numeric and non-numeric columns
numeric_cols = data_train.select_dtypes(include=[np.number]).columns
non_numeric_cols = data_train.select_dtypes(exclude=[np.number]).columns

# Handle missing values for numeric columns
data_train[numeric_cols] = data_train[numeric_cols].fillna(data_train[numeric_cols].mean())
data_test[numeric_cols] = data_test[numeric_cols].fillna(data_train[numeric_cols].mean())  # Fill with training data mean

# Concatenate the numeric and non-numeric columns back together
data_train = pd.concat([data_train[numeric_cols], data_train[non_numeric_cols]], axis=1)
data_test = pd.concat([data_test[numeric_cols], data_test[non_numeric_cols]], axis=1)

# Encode categorical variables (create dummy variables)
categorical_cols = data_train.select_dtypes(include=['object']).columns
data_train = pd.get_dummies(data_train, columns=categorical_cols, drop_first=True)
data_test = pd.get_dummies(data_test, columns=categorical_cols, drop_first=True)

# Make sure both training and test datasets have the same columns after encoding
common_columns = list(set(data_train.columns) & set(data_test.columns))
data_train = data_train[common_columns]
data_test = data_test[common_columns]

# Split the data into features and target
X_train = data_train.drop("Car_Price", axis=1)
y_train = data_train["Car_Price"]
X_test = data_test.drop("Car_Price", axis=1)
y_test = data_test["Car_Price"]

# Model Selection - Initialize models
linear_regression = LinearRegression()
lasso = Lasso(alpha=0.01)  
ridge = Ridge(alpha=1.0)  
random_forest = RandomForestRegressor(n_estimators=100)
gradient_boosting = GradientBoostingRegressor(n_estimators=100)

# Train and evaluate models
models = [linear_regression, lasso, ridge, random_forest, gradient_boosting]

for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"Model: {model.__class__.__name__}")
    print(f"Mean Squared Error: {mse}")
    print(f"R-squared: {r2}")
    print("\n")

# Ensemble Models
ensemble_model = GradientBoostingRegressor(n_estimators=100)  

ensemble_model.fit(X_train, y_train)
y_ensemble_pred = ensemble_model.predict(X_test)

# Evaluate the ensemble model
mse_ensemble = mean_squared_error(y_test, y_ensemble_pred)
r2_ensemble = r2_score(y_test, y_ensemble_pred)
print("Ensemble Model")
print(f"Mean Squared Error: {mse_ensemble}")
print(f"R-squared: {r2_ensemble}")

```
